## Recursion and Backtracking

Definir de forma recursiva el factorial de un número.

```python
def factorial (n):
    if n<= 0: 
        return 1
    return n* factorial(n-1)
```

```python
print (f'El factorial de 10 es: {factorial(10)}')
```

An example of backtracking is given next. Here, we have used a recursive approach to generate all the possible arrangements of a given string, s, of a given length, n:

```python
def bitStr (str, n):
    if (n == 1): return str
    return [digit + bits for digit in bitStr(str, 1) for bits in bitStr(str, n-1)]
```

```python
value = bitStr('abc', 3)
print (f'{value}')
```

Implementación del algoritmo de Karatsuba para calcular la multiplicación de números

```python
from math import log10
import math

def karatsuba (x,y): 
    # Establish the base case
    if (x <10 or y <10): return x * y

    # Set n, the number of digits in the highest input number
    n = max (int (log10 (x)+1), int (log10 (y)+1))

    # round up n/2
    n_2 = int (math.ceil(n/2.0))

    #add 1 if n is odd
    n = n if n % 2 == 0 else n + 1

    # Splits the input numbers
    a,b = divmod (x, 10**n_2)
    c,d = divmod (y, 10**n_2)

    #Application of the third recursive steps
    ac = karatsuba(a,c)
    bd = karatsuba(b,d)
    ad_bc = karatsuba((a+b), (c+d))- ac - bd

    # Finally, we could perform the multiplication
    return (((10**n)*ac) + bd + ((10**n_2)*(ad_bc)))

```

```python
t= karatsuba(1234, 3456)
print (t)
```
